### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()

# 1. Análisis  facturación total de cada tienda y a partir de la información recopilada, observar cuál tienda vende mas.
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
print(f'Facturación total de la Tienda 1 es de {ingreso_tienda1}')
print(f'Facturación total de la Tienda 2 es de {ingreso_tienda2}')
print(f'Facturación total de la Tienda 3 es de {ingreso_tienda3}')
print(f'Facturación total de la Tienda 4 es de {ingreso_tienda4}')


if ingreso_tienda1 < ingreso_tienda2 and ingreso_tienda1 < ingreso_tienda3 and ingreso_tienda1 < ingreso_tienda4:
  print(f'La tienda 1 vende menos con un valor de {ingreso_tienda1}')
elif ingreso_tienda2 < ingreso_tienda1 and ingreso_tienda2 < ingreso_tienda3 and ingreso_tienda2 < ingreso_tienda4:
  print(f'La tienda 2 vende menos, con un valor de {ingreso_tienda2}')
elif ingreso_tienda3 < ingreso_tienda1 and ingreso_tienda3 < ingreso_tienda2 and ingreso_tienda3 < ingreso_tienda4:
  print(f'La tienda 3 vende menos con un valor de {ingreso_tienda3}')
else:
  print(f'La tienda 4 vende menos con un valor de {ingreso_tienda4}')





```

#1. Análisis de facturación



```python
# Visualización del comportamiento de las ventas en las tiendas

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(tiendas, [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4])
plt.title('Facturación por tienda')

plt.xlabel('Tiendas')
plt.ylabel('Facturación')

plt.show()
```

# 5. Envío promedio por tienda

```python

costoEnvioTienda = tienda['Costo de envío'].mean()
costoEnvioTienda2 = tienda2['Costo de envío'].mean()
costoEnvioTienda3 = tienda3['Costo de envío'].mean()
costoEnvioTienda4 = tienda4['Costo de envío'].mean()

print('Costo de envío de la tienda 1: ', costoEnvioTienda)
print('Costo de envío de la tienda 2: ', costoEnvioTienda2)
print('Costo de envío de la tienda 3: ', costoEnvioTienda3)
print('Costo de envío de la tienda 4: ', costoEnvioTienda4)


listaTiendas = ['tienda 1', 'tienda 2', 'tienda 3', 'tienda 4']
costosEnvios = [costoEnvioTienda, costoEnvioTienda2, costoEnvioTienda3, costoEnvioTienda4]
colores = ['#533B4D', '#7F55B1', '#8ACCD5', '#644A07']
plt.figure(figsize=(8, 5))
barras = plt.bar(listaTiendas, costosEnvios, width=0.6, color=colores)
plt.title('Costos de envíos promedio por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Promedio de envíos')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Agrega etiquetas as barras
for barra in barras:
    valory = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, valory, f"${valory:,.0f}", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
